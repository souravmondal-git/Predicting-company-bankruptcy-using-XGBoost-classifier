```python
y=df['Bankrupt?']
```

```python
X = df.drop(['Bankrupt?'], axis=1)
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=1)
```

```python
from hpsklearn import HyperoptEstimator
from hyperopt import hp
from hyperopt import STATUS_OK, Trials, fmin, tpe
from xgboost import XGBClassifier
```

```python
from sklearn.metrics import accuracy_score
```

```python
space={'max_depth': hp.quniform("max_depth", 3, 9, 1),
        'gamma': hp.uniform ('gamma', 1,9),
        'alpha' : hp.quniform('alpha', 40,100,1),
        'lambda' : hp.uniform('lambda', 0,1),
        'colsample_bytree' : hp.uniform('colsample_bytree', 0.5,1),
        'min_child_weight' : hp.quniform('min_child_weight', 0, 10, 1),
        'n_estimators': hp.uniformint('n_estimators',200,300),
        'seed': 0
    }
```

```python
def objective(space):
    clf=XGBClassifier(
                    n_estimators =space['n_estimators'], max_depth = int(space['max_depth']), gamma = space['gamma'],
                    alpha = int(space['alpha']),min_child_weight=int(space['min_child_weight']),
                    colsample_bytree=int(space['colsample_bytree']))
    
    evaluation = [( X_train, y_train), ( X_test, y_test)]
    
    clf.fit(X_train, y_train,
            eval_set=evaluation, eval_metric="auc",
            early_stopping_rounds=10,verbose=False)
    

    pred = clf.predict(X_test)
    accuracy = accuracy_score(y_test, pred>0.5)
    print ("SCORE:", accuracy)
    return {'loss': -accuracy, 'status': STATUS_OK }
```

```python
trials = Trials()

best_hyperparams = fmin(fn = objective,
                        space = space,
                        algo = tpe.suggest,
                        max_evals = 100,
                        trials = trials)
```

```python
print('best hyp params:',best_hyperparams)
```

```python
model = XGBClassifier(parameters=best_hyperparams)
```

```python
model.fit(X_train, y_train)
```

```python
y_pred = model.predict(X_test)
```

```python
pred_prob = model.predict_proba(X_test)
```

```python
from sklearn.metrics import roc_curve, roc_auc_score
```

```python
fpr, tpr, thrshld = roc_curve(y_test, pred_prob[:,1])
```

```python
plt.figure(figsize=(8,8))
plt.title('ROC Curve')
plt.plot(fpr, tpr)
plt.plot([0,1],ls='--')
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.show()
```

```python
print('AUC score :', roc_auc_score(y_test, pred_prob[:,1]))
```
